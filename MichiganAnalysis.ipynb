```python
# Replace Unranked seasons with the Value of 40. 40= Unranked

finaldf=df.fillna(40) 
```

```python
#Average wins per season by coach
WinsByCoach = finaldf.groupby('Coach')['W'].mean()
```

```python
WinsByCoach.head(50)
```

```python
PositionByCoach =finaldf.groupby('Coach')['AP Post'].mean()
```

```python
PositionByCoach.head(10)
```

```python
WorstFinalPosition =finaldf.groupby('Coach')['AP Post'].max()
```

```python
WorstFinalPosition.head(10)
```

```python
#Best Finish in AP Poll
BestFinalPosition =finaldf.groupby("Coach")['AP Post'].min()
```

```python
BestFinalPosition.head(10)
```

```python
finaldf['AP Post']
```

```python
BestFinalPosition.head(10)
```

```python
# Average Wins Per Coach
%matplotlib inline

import matplotlib.pyplot as plt

WinsByCoach.plot.bar(color=['blue'])
plt.title('MichiganWins')
plt.xlabel('Coach')
plt.ylabel('W')
plt.legend(['Count of rows'])
```

```python
# Worst Final AP Poll by Coach. 40= unranked
WorstFinalPosition.plot.bar(color=['blue'])
plt.title('Worst Final Position As Coach')
plt.xlabel('Coach')
plt.ylabel('AP Post')
```

```python
#Best Final Positon by final AP poll. LLoyd Carr's first place finish was only national championship during these time period
BestFinalPosition.plot.bar(color=['black'])
plt.title('Best Final Position')
plt.xlabel('Coach')
plt.ylabel('AP Post')
```

```python
# Suptract PreSeason AP Poll by end of season AP Poll Standing to see how they finihsed the season in comparison to the expectations of the beginning of the season
finaldf['Expected']=finaldf['AP Pre']-finaldf['AP Post']
```

```python
FinalExpected=finaldf.groupby('Coach')['Expected'].mean()
```

```python
FinalExpected.head(10)
```

```python
FinalExpected.plot.bar(color=['Red'])
plt.title('Expectations')
plt.xlabel('Coach')
#plt.ylabel('Expectations')
```

```python
# Average Ranking by coach
PositionByCoach.plot.bar(color=['green'])
plt.title('Average Ranking By Coach')
plt.xlabel('Coach')
plt.ylabel('Ranking')
```

```python
PositionByCoach.plot.line(color=['green'])
plt.title('Average Ranking By Coach')
plt.xlabel('Coach')
plt.ylabel('Ranking')
```
